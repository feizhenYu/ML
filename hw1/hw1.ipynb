```python
import sys
import numpy as np
import pandas as pd

data = pd.read_csv('train.csv', encoding='big5')
```

```python
data = data.iloc[:, 3:]
print(data)
```

```python
data[data=='NR'] = 0
rawdata = data.to_numpy()
print(rawdata)
```

```python
month_data = {}
for month in range(12):
    sample = np.empty([18, 480])
    for day in range(20):
        sample[:, day * 24: (day + 1) * 24] = rawdata[(month * 20 + day) * 18: (month * 20 + day + 1) * 18, :]
    month_data[month] = sample
print(month_data[0])
```

```python
x = np.empty([12 * 471, 18 * 9], dtype=float)
y = np.empty([12 * 471, 1], dtype=float)
for month in range(12):
    for hour in range(471):
        x[month * 471 + hour, :] = month_data[month][:, hour : hour + 9].reshape(1, -1)
        y[month * 471 + hour, 0] = month_data[month][9, hour + 9]
print(x)
print(y)        
```

```python
mean_x = np.mean(x, axis=0)
std_x = np.std(x, axis=0)
for j in range(len(x[0])):
    if std_x[j] != 0:
        for i in range(len(x)):
            x[i][j] = (x[i][j] - mean_x[j])/std_x[j]
print(x)            
```

```python
import math
x_train_set = x[: math.floor(len(x) * 0.8), :]
y_train_set = y[: math.floor(len(x) * 0.8), :]
x_validation_set = x[math.floor(len(x) * 0.8):, :]
y_validation_set = y[math.floor(len(x) * 0.8):, :]
print(len(x_train_set))
print(len(y_train_set))
print(len(x_validation_set))
print(len(y_validation_set))
```

```python
dim = 18 * 9 + 1
w = np.zeros([dim, 1])
x = np.concatenate((np.ones([12 * 471, 1]), x), axis = 1).astype(float)
lr = 100
iter_time = 1000
adagrad = np.zeros([dim, 1])
eps = 0.00000001
for t in range(iter_time):
    loss = np.sqrt(np.sum(np.power(np.dot(x, w)-y, 2))/471/12)
    if (t % 100 == 0):
        print(str(t),':',str(loss))
    gradient = np.dot(x.transpose(), np.dot(x,w)-y) * 2
    adagrad += gradient ** 2
    w = w - lr * gradient/np.sqrt(adagrad + eps)
np.save('weight.npy', w)
print(w)
```

```python
testdata = pd.read_csv('./test.csv', header = None, encoding = 'big5')
test_data = testdata.iloc[:, 2:]
test_data[test_data == 'NR'] = 0
test_data = test_data.to_numpy()
test_x = np.empty([240, 18*9], dtype = float)
for i in range(240):
    test_x[i, :] = test_data[18 * i: 18* (i + 1), :].reshape(1, -1)
for i in range(len(test_x)):
    for j in range(len(test_x[0])):
        if std_x[j] != 0:
            test_x[i][j] = (test_x[i][j] - mean_x[j]) / std_x[j]
test_x = np.concatenate((np.ones([240, 1]), test_x), axis = 1).astype(float)
test_x

w = np.load('weight.npy')
ans_y = np.dot(test_x, w)
ans_y

import csv
with open('submit.csv', mode='w', newline='') as submit_file:
    csv_writer = csv.writer(submit_file)
    header = ['id', 'value']
    print(header)
    csv_writer.writerow(header)
    for i in range(240):
        row = ['id_' + str(i), ans_y[i][0]]
        csv_writer.writerow(row)
        print(row)
```
